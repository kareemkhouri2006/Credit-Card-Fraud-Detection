```python
import pandas as pd
```

read files

```python
aisles=pd.read_csv("../data/aisles.csv")
departments=pd.read_csv("../data/departments.csv")
order_products_prior=pd.read_csv("../data/order_products__prior.csv")
order_products_train=pd.read_csv("../data/order_products__train.csv")
orders=pd.read_csv("../data/orders.csv")
products=pd.read_csv("../data/products.csv")
```

memory reduction

```python
def reduce_memory(df):
    for col in df.columns:

        # skip IDs
        if col in ["aisle_id", "department_id", "order_id", "product_id", "user_id"]:
            continue

        col_type = df[col].dtype

        # integers
        if pd.api.types.is_integer_dtype(col_type):
            df[col] = pd.to_numeric(df[col], downcast="integer")

        # floats
        elif pd.api.types.is_float_dtype(col_type):
            df[col] = pd.to_numeric(df[col], downcast="float")

        # objects → category ONLY if repeated values
        elif col_type == "object":
            if df[col].nunique() / len(df) < 0.5:
                df[col] = df[col].astype("category")

    return df      

aisles=reduce_memory(aisles)
departments=reduce_memory(departments)
order_products_prior=reduce_memory(order_products_prior)
order_products_train=reduce_memory(order_products_train)
orders=reduce_memory(orders)
products=reduce_memory(products)
```

```python
df_prior= (order_products_prior
.merge(products,on="product_id",how="left")
.merge(orders,on="order_id",how="left")
.merge(aisles,on="aisle_id",how="left")
.merge(departments,on="department_id",how="left"))
```

```python
df_train= (order_products_train
.merge(products,on="product_id",how="left")
.merge(orders,on="order_id",how="left")
.merge(aisles,on="aisle_id",how="left")
.merge(departments,on="department_id",how="left"))
```
